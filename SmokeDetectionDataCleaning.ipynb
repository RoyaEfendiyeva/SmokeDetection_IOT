```python
df_cleaned = df.dropna()   #------->option 1
#df_cleaned['Temperature[C]'].fillna(df_cleaned['Temperature[C]'].mean(), inplace=True)   #------->option 2
```

```python
df_cleaned = df_cleaned.drop_duplicates()
```

```python
df_cleaned
```

```python
df_cleaned['UTC'] = pd.to_datetime(df_cleaned['UTC'], unit='s')
```

```python
alarm_triggered = df_cleaned[df_cleaned['Fire Alarm'] == 1]
non_alarm_data = df_cleaned[df_cleaned['Fire Alarm'] == 0]
for column in ['Temperature[C]', 'Humidity[%]', 'TVOC[ppb]', 'eCO2[ppm]', 'Raw H2', 'Raw Ethanol', 'Pressure[hPa]', 'PM1.0', 'PM2.5', 'NC0.5', 'NC1.0', 'NC2.5']:
    z_scores = stats.zscore(non_alarm_data[column])  #Z-balı məlumat nöqtəsinin ortadan nə qədər uzaq olduğunu standartlaşdırır.
    abs_z_scores = np.abs(z_scores)
    filtered_entries = (abs_z_scores < 3)  #z-balı 3-dən çox olan dəyərlər kənar hesab olunur.
    non_alarm_data = non_alarm_data[filtered_entries]
```

```python
df_cleaned=pd.concat([non_alarm_data,alarm_triggered],ignore_index=True)
```

```python
cleaned_file_path = r'C:\Users\HP\Downloads\CleanedSmokeDetection_iot.csv'
df_cleaned.to_csv(cleaned_file_path, index=False)
print("\nCleaned dataset saved to:", cleaned_file_path)
```
