```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python

# CSV faylını yükləyirik
df = pd.read_csv(r'C:\Users\HP\Downloads\CleanedSmokeDetection_iot.csv')

# UTC sütununu tarix-zaman formatına çeviririk (əgər lazım olsa)
df['UTC'] = pd.to_datetime(df['UTC'])

# UTC sütununu indeks olaraq təyin edirik
df.set_index('UTC', inplace=True)

# Saatlıq intervalla aqreqasiyanı həyata keçiririk
aggregated_df = df.resample('h').mean().reset_index()
```

```python


# Qrafik yaradılması
plt.figure(figsize=(14, 7))
plt.plot(aggregated_df['UTC'], aggregated_df['Temperature[C]'], marker='o', linestyle='-')
plt.title('Temperature Variation Over Time (Hourly Aggregated)')
plt.xlabel('Timestamp')
plt.ylabel('Temperature [C]')
plt.grid(True)
plt.tight_layout()
plt.show()

```
