```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Load cleaned data from CSV file
df = pd.read_csv(r'C:\Users\HP\Downloads\CleanedSmokeDetection_iot.csv')
```

```python


# Select a column for example, e.g., 'Temperature[C]'
data = df['Temperature[C]']

# Calculate quartile values
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Calculate whisker boundaries
lower_whisker= Q1 - 1.5 * IQR   #min
upper_whisker = Q3 + 1.5 * IQR  #max

print(f'Q1 (25th percentile): {Q1}')
print(f'Q3 (75th percentile): {Q3}')
print(f'IQR: {IQR}')
print(f'Lower whisker boundary: {lower_whisker}')
print(f'Upper whisker boundary: {upper_whisker}')

```

```python

# Set the figure size
plt.figure(figsize=(6,12))

# Create a boxplot with additional settings
sns.boxplot(data=df[['PM1.0', 'PM2.5', 'eCO2[ppm]', 'TVOC[ppb]']],
            boxprops=dict(alpha=.7, color='black'),  # Color and transparency of the boxes
            whiskerprops=dict(color='blue', linestyle='--'),  # Style of the whiskers
            capprops=dict(color='green'),  # Color of the caps
            medianprops=dict(color='red', linewidth=0.5),  # Style of the median line
            flierprops=dict(marker='o', color='orange', alpha=1))  # Style of the outliers

# Set the title of the plot
plt.title('Boxplot of Various Variables')

# Set the label for the Y axis
plt.ylabel('Values')

# Add a grid for better readability
plt.grid(True)

# Display the plot
plt.show()

```
