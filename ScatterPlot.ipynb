```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#CLEANED DATA
df = pd.read_csv(r'C:\Users\HP\Downloads\CleanedSmokeDetection_iot.csv')
```

```python
# Adjust figure size if needed
plt.figure(figsize=(8, 6))   #Qrafik ölçüsünü 8x6 düym olaraq təyin edir

# Scatter plot with smaller markers and transparency
sns.scatterplot(x='Temperature[C]', y='Humidity[%]', data=df, s=50, alpha=0.5,color='green')   #s=50 Marker ölçüsünü 50-yə təyin edir.

plt.title('Temperature vs Humidity')
plt.xlabel('Temperature [C]')
plt.ylabel('Humidity [%]')
plt.grid(True)   # Qrafikdə tor xətlərini göstərir.
plt.tight_layout() #Etiketlərin və başlıqların düzgün görünməsi üçün diaqram elementlərinin düzülməsinə ciddi şəkildə düzəliş etməyə imkan verir.
plt.show()
```
