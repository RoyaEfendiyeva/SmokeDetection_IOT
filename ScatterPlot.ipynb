```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#CLEANED DATA
df = pd.read_csv(r'C:\Users\HP\Downloads\CleanedSmokeDetection_iot.csv')
```

```python
# Adjust figure size if needed
plt.figure(figsize=(8, 6))   #Qrafik lsn 8x6 dym olaraq tyin edir

# Scatter plot with smaller markers and transparency
sns.scatterplot(x='Temperature[C]', y='Humidity[%]', data=df, s=50, alpha=0.5,color='green')   #s=50 Marker lsn 50-y tyin edir.

plt.title('Temperature vs Humidity')
plt.xlabel('Temperature [C]')
plt.ylabel('Humidity [%]')
plt.grid(True)   # Qrafikd tor xtlrini gstrir.
plt.tight_layout() #Etiketlrin v balqlarn dzgn grnmsi n diaqram elementlrinin dzlmsin ciddi kild dzli etmy imkan verir.
plt.show()
```
